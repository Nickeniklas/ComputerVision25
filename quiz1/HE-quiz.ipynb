### Show image histograms

using **matplotlib pyplot**

```python
# läs in bilderna
bild1 = cv.imread("bild1.jpg", cv.IMREAD_GRAYSCALE)
bild2 = cv.imread("bild2.jpg", cv.IMREAD_GRAYSCALE)
bild3 = cv.imread("bild3.jpg", cv.IMREAD_GRAYSCALE)
bild4 = cv.imread("bild4.jpg", cv.IMREAD_GRAYSCALE)
bild5 = cv.imread("bild5.jpg", cv.IMREAD_GRAYSCALE)
```

```python
# plot histograms using pyplot
plt.hist(bild1.flatten(),256,[0,256], color = 'r')
plt.title("bild1")
#plt.show()
```

```python
# plot all histograms
bilder = [bild1, bild2, bild3, bild4, bild5]
for i, bild in enumerate(bilder):
    plt.hist(bild.flatten(),256,[0,256], color = 'r')
    plt.title(f"bild{i +1}")
    plt.show()
```

### enhance x-ray image

using **cv.equalizeHist()**

x-ray.jpg

```python
xrayImg = cv.imread("x-ray.jpg", cv.IMREAD_GRAYSCALE)

cv.imshow("original x-ray", xrayImg)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# enhance
enhancedXray = cv.equalizeHist(xrayImg)

cv.imshow("original x-ray", enhancedXray)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# show new histogram
plt.hist(enhancedXray.flatten(),256,[0,256], color = 'r')
plt.title("new histogram")
plt.show()
```
