```python
# imports and installs
%pip install matplotlib
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# read images
building = cv.imread("building.jpg")
grayBuilding = cv.cvtColor(building, cv.COLOR_BGR2GRAY)
```

```python
dst = cv.cornerHarris(grayBuilding, 2, 3, 0.04)
dst = cv.dilate(dst, None)

threshold = 0.0001 * dst.max()
building[dst > threshold] = [255, 0, 0]
fig = plt.figure()
fig.set_size_inches(8, 8)
plt.imshow(cv.cvtColor(building, cv.COLOR_BGR2RGB))
```

Decreasing the **Threshold** size, enabled it to detect the top most corners of the building.
