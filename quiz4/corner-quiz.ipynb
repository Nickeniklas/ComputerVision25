```python
# imports and installs
%pip install matplotlib
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# read images
building = cv.imread("building.jpg")
grayBuilding = cv.cvtColor(building, cv.COLOR_BGR2GRAY)
```

```python
dst = cv.cornerHarris(grayBuilding, 2, 3, 0.4)
dst = cv.dilate(dst, None)

threshold = 0.01 * dst.max()
colorBuilding = cv.cvtColor(grayBuilding, cv.COLOR_GRAY2BGR)
colorBuilding[dst > threshold] = [255, 0, 0]
fig = plt.figure()
fig.set_size_inches(8, 8)
plt.imshow(cv.cvtColor(colorBuilding, cv.COLOR_BGR2RGB))
```
